```python
import spacy

nlp = spacy.load("en_core_web_sm")
sentence = "The quick brown fox jumps over the lazy dog."
doc = nlp(sentence)
lemmas = [token.lemma_ for token in doc]
print("spaCy Lemmatizer:", lemmas)
```
